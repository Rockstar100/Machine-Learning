```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
df = pd.read_csv('csvFile.csv')
df
```

```python
cg = pd.get_dummies(df.city)
cg
```

```python
df_cg = pd.concat([df,cg], axis='columns')
df_cg
```

```python
df_final = df_cg.drop(['city','Karnal'], axis = 'columns')
df_final
```

```python
y = df_cg.price
y
```

```python
x = df_final.drop('price', axis='columns')
x
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
model.predict([[3000,0,0]])
```
